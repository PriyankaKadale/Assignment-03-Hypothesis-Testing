## Hypothesis Testing Exercise

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
buyer_ratio = pd.read_csv('BuyerRatio.csv')
buyer_ratio.head()
```

```python
ratio = buyer_ratio.drop('Observed Values', axis=1)
ratio
```

```python
from scipy.stats import chi2_contingency
stat, pvalue, dof, expected = chi2_contingency(ratio)
```

```python
stat
```

```python
pvalue
```

```python
if pvalue<0.05:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```
